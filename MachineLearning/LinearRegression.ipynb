```python
import numpy as np
import matplotlib.pyplot as plt

N = data.shape[0]
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)

x = np.hstack((np.ones((N, 1)), x))  # Thêm cột bias để x có kích thước (N, 2)
w = np.array([0.0, 1.0]).reshape(-1, 1)  # Đảm bảo w có kiểu float

numOfIteration = 100
cost = np.zeros((numOfIteration, 1))
learning_rate = 0.000000001

for i in range(1, numOfIteration):
    r = np.dot(x, w) - y
    cost[i] = 0.5 * np.sum(r * r)
    w[0] -= learning_rate * np.sum(r)
    w[1] -= learning_rate * np.sum(np.multiply(r, x[:, 1].reshape(-1, 1)))
    print(cost[i])

predict = np.dot(x, w)
plt.plot((x[0][1], x[N-1][1]), (predict[0], predict[N-1]), 'r')
plt.show()
```

b) Code python thuật toán gradient descent để tìm 
giá trị nhỏ nhất của hàm f(x) = x^2 + 2x + 5

```python
import numpy as np

def f(x):
    return x ** 2 + 2 * x + 5
def FDaoHam(x):
    return 2 * x + 2
def GradientDescent(xo = 0, sobuoclap = 1000, learning_rate = 0.000001):
    x = xo
    for i in range(sobuoclap):
        x = x - learning_rate*FDaoHam(x)
    return x, f(x)
min_x, fx = GradientDescent()

print('Gia tri nho nhat: {}, tai {}'.format(min_x, fx))
```

```python
X = df[['DienTich_m2', 'SoPhongNgu']].values
Y = data['Gia_Tri_VND']
```

```python
N = X.shape[0]
X = np.hstack((np.ones((N, 1)), X))
w = np.zeros((X.shape[1], 1))

soVongLap = 1000
learning_rate = 0.01

for i in range(soVongLap):
    y_pred = np.dot(X, w)
    r = y_pred - Y
    w[0] -= learning_rate * np.sum(r)
    w[1] -= learning_rate * np.sum(X[:, 1].reshape(-1, 1)*r)
    w[2] -= learning_rate * np.sum(X[:, 2].reshape(-1, 1) * r)


predict = np.dot(X, w)

# Vẽ biểu đồ
plt.scatter(df['DienTich_m2'], Y, label='Giá thực tế')  # Vẽ giá thực tế cho diện tích
plt.scatter(df['SoPhongNgu'], Y, label='Giá thực tế theo số phòng ngủ')  # Vẽ giá thực tế theo số phòng ngủ

# Vẽ giá dự đoán
plt.scatter(df['DienTich_m2'], predict, color='red', label='Dự đoán theo Diện tích', alpha=0.5)

plt.xlabel('Diện tích (m²) và Số phòng ngủ')
plt.ylabel('Giá nhà')
plt.title('Dự đoán Giá Nhà dựa trên Diện Tích và Số Phòng Ngủ')
plt.legend()
plt.show()

```
